7. Fit the full factorial ANOVA model. Check the assumption of normality using a testing method and a visual method. Does the assumption hold? Why or why not?

The points in the QQ-Plot should be examined to determines whether there are strong deviations from the diagnal reference line. If only minor deviations are present this provides evidence in favor of the normality assumptions.
Additionally, if the p-value associated with the Shapiro-Wilk test is greater than the commonly used α =  threshold, we fail to reject the null hypothesis of normality . Thus the data would support the assumption that the residuals are approximately normally distributed.

```python
from numpy.random import MT19937

## Question 7 Code ##

m1 = ols("Satisfaction ~ C(Protein) * C(Salsa) * C(Potatoes)",
         data=burritos).fit()

## Evaluate Normality ##

## QQ Plot ##

sm.qqplot(m1.resid, line='s')
plt.title("QQ Plot of Residuals")
plt.show()

## Shapiro-Wilk Test ##

shapiro_test = stats.shapiro(m1.resid)

print("Shapiro-Wilk Test Statistic:", shapiro_test.statistic)
print("Shapiro-Wilk Test p-value:", shapiro_test.pvalue)

if shapiro_test.pvalue > 0.05:
    print("Residuals are approximately normally distributed (fail to reject H0).")
else:
    print("Residuals are not normally distributed (reject H0).")
```

A fractional factorial design is appropriate in this situation because it allows us to evaluate multiple factors while reducing the number of experimental runs required. This saves time, labor, and operational costs.

7. Assuming a defining relation of $I=ABC$, specify the aliasing structure of the design. What does this mean for the interpretation of the main effects and interactions?

The defining relation for this fractional factorial design is I = ABC. This means that the main effects are aliased with the two-way interaction effects. Specifically:

A is aliased with BC
B is aliased with AC
C is aliased with AB

As a result, we cannot independently separate the main effects from the corresponding two-way interaction effects. Therefore, when interpreting the results, any observed main effect may partially reflect the associated interaction effect.


```python
## Question 7 Code ##

## Aliasing Structure ##

print("Defining Relation: I = ABC")

print("Alias Structure:")

print("A = BC")
print("B = AC")
print("C = AB")
```

```python
## Question 8 Code ##

## Fit ANOVA Model ##

m1 = ols(
    "Sales ~ C(Location) + C(Timing) + C(Marketing)",
    data=sales
).fit()

## Evaluate Normality ##

## QQ Plot ##

sm.qqplot(m1.resid, line='s')

plt.title("QQ Plot of Residuals")

plt.show()

## Shapiro-Wilk Test ##

shapiro_test = stats.shapiro(m1.resid)

print("Shapiro-Wilk Test Statistic:",
      shapiro_test.statistic)

print("Shapiro-Wilk Test p-value:",
      shapiro_test.pvalue)

if shapiro_test.pvalue > 0.05:
    print(
        "Residuals are approximately normally distributed (fail to reject H0)."
    )
else:
    print(
        "Residuals are not normally distributed (reject H0)."
    )
```
